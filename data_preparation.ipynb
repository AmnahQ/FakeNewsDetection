## Data Cleaning

1. Removed URLs
2. Removed User Mentions
3. Removed all numbers
4. Removed punctuations and extra spaces

```python
import string
import re

df1['text'] = df1['text'].apply(lambda x:re.sub(r'http\S+', '', x))
df1['text'] = df1['text'].apply(lambda x:re.sub(r'@\S+ ', '', x))
df1['text'] = df1['text'].apply(lambda x:''.join(i for i in x if not i.isdigit()))
table = str.maketrans(string.punctuation, ' '*len(string.punctuation))
df1['text'] = df1['text'].str.translate(table)
df1['text'] = df1['text'].str.replace(' +', ' ')
df1['text'] = df1['text'].str.lower()
df1['text'] = df1['text'].str.strip()

display(df1.head())
```
